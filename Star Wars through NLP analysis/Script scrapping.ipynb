```python
import pandas as pd
import numpy as np
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
from selenium.webdriver.support.ui import Select
```

```python
driver = webdriver.Firefox(executable_path=r'geckodriver.exe')
```

```python
link='https://www.imsdb.com/scripts/Star-Wars-The-Force-Awakens.html'
```

```python
driver.get(link)
```

```python
element=driver.find_element_by_xpath("//td[@class='scrtext']")
```

```python
script=element.text
```

```python
script=script.split('\n')
```

```python
def upper_line(line):
    
    line_sp=line.split(" ")
    line_sp=[x for x in line_sp if len(x)>0]
    if len(line_sp)>0:
        return np.sum([x.isupper() for x in line_sp])/len(line_sp)
    else: 
        return 2   
```

```python
def remove_brackets(line):
    start = line.find( '(' )
    end = line.find( ')' )
    if start != -1 and end != -1:
        return line[:start]+line[end+1:]
    else: return line
```

```python
def extract_words(i):    
    words=[]
    k=1
    while upper_line(script[i+k])<0.5:
        line_sp=script[i+k].split(" ")
        line_sp=[x for x in line_sp if len(x)>0]
        words.append(' '.join(line_sp))
        k+=1
    
    words=' '.join(words)
    words=remove_brackets(words)
    
    return words
```

```python
characters=['KYLO REN', 'POE', 'REY', 'LUKE', 'FINN', 'HAN', 'LEIA', "LARMA D'ACY", 'AMILYN HOLDO', 'SNOKE', 'GENERAL HUX', 'SNOKE', 'CAPTAIN PHASMA', 'MAZ']
```

```python
df=pd.DataFrame()
```

```python
for character in characters:
    replics=[extract_words(i) for i in range(len(script)) if (character in script[i])]
    replics=[x for x in replics if x!='']
    df=df.append(pd.DataFrame(list(zip([character]*len(replics), replics))))
```

```python
df.to_csv('force awakens.csv')
```
